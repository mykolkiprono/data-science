```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
x = np.random.randn(500)
y = np.random.randn(500)
```

```python
%pylab inline
plt.scatter(x,y)
plt.ylim(-2,2)
plt.xlim(-2,2)
plt.title("scatter")
plt.show()
```

```python
from matplotlib.backends.backend_agg import new_figure_manager
```

```python
fig = new_figure_manager(1).canvas.figure
ax = fig.add_subplot(211)
ax.scatter(x,y)
ax.set_ylim(-2,2)
ax.set_xlim(-2,2)
ax.set_title("scatter")

ax2 = fig.add_subplot(212)
ax2.scatter(x,y)
ax.set_ylim(-2,2)
ax.set_xlim(-2,2)
ax2.set_title("scatter")
display(fig)
# plt.show()
```

```python
t = np.linspace(0,10,300)
x  = np.sin(2*pi*0.1*t**2)
```

```python
fig = new_figure_manager(1).canvas.figure

ax = fig.add_subplot(111)
ax.step(t,x)
ax.autoscale_view()
display(fig)
```

```python
from matplotlib.lines import Line2D

fig = plt.figure()
ax = fig.add_subplot(111)
line = Line2D(x,y)
ax.add_artist(line)
plt.show()
```

```python
ax.autoscale_view()
ax.set_aspect(1.0, 'datalim')
line.set_color('red')
line.set_linestyle('dashed')
plt.draw()
display(fig)

# for reference
# add_artist()
# add_line()
# add_collection()
# add_patch()
```

```python
from matplotlib.collections import CircleCollection
```

```python
x = x.reshape(-1,1)
y = y.reshape(-1,1)
```

```python
fig = plt.figure()
ax = fig.add_subplot(111)
sizes = np.empty_like(x)
sizes.fill(20)
col1 = CircleCollection(sizes, offsets = (x,y), transOffset = ax.transData, edgecolor = 'red', facecolor = 'cyan')
ax.add_collection(col1)
ax.autoscale_view()
ax.draw()
```

```python
from matplotlib.patches import Rectangle, Circle
```

```python
circe = Circle((0,0), radius = 0.001, color = 'black',alpha = 0.8, )
ax.add_patch(circe)

rect = Rectangle((0.002,0.01), width = 0.005, height = 0.005, alpha = 0.8, color = 'red')
ax.add_patch(rect)
plt.draw()
display(fig)
# Patches : cicles, rectangles, ellipses, Arc, arrow, polygon, 
```
