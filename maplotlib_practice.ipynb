```python
from matplotlib import pylab
```

```python
import numpy as np
```

```python
x=np.linspace(0,10,25)
y=x*x+5
print(x)
print(y)
```

```python
pylab.plot(x,y,'r')
#pylab.tittle("DISTANCE ACCELERRATION GRAPH")
pylab.xlabel("distance")
pylab.ylabel("acceleration")
pylab.show()
```

```python
pylab.subplot(1,2,1)
pylab.plot(x,y,'r--')
pylab.subplot(1,2,2)
pylab.plot(x,y,'g*-')
pylab.show()
```

```python
import matplotlib.pyplot as plt
```

```python
fig=plt.figure()
axis=fig.add_axes([0.5,0.5,0.8,0.8])#left right height and width
axis.plot(x,y,'r')
plt.show()
```

```python
#ploting subgraphs using pyplot
fig,axes=plt.subplots(nrows=1,ncols=4) #submap is of 1 row and 2 columns
for ax in axes:
    ax.plot(x,y,'r')
```

```python
#we can also draw a picture or graph inside another graph
fig=plt.figure()
#control the left,bottom,width and height of the canvas
axis1=fig.add_axes([0.1,0.1,0.8,0.8])
axis2=fig.add_axes([0.2,0.6,0.3,0.2])
axis1.plot(x,y,'r')#big axes
axis2.plot(y,x,'g')#small canvas
plt.show()
```

```python
#in matplotlib you can set some ather properties  such as line color,trnsparency and more
fig,axes=plt.subplots(dpi=150)
axes.plot(x,x**2,color='red',alpha=.5)
axes.plot(x,x**.5,color='blue',alpha=.5)
axes.plot(x,x+3,color='green',alpha=.5)
plt.show()
```

```python
fig,ax=plt.subplots(dpi=150)
#line width
ax.plot(x,x+1,color='blue',linewidth=0.25)
ax.plot(x,x+2,color='blue',linewidth=0.8)
ax.plot(x,x+3,color='blue',lw=0.2)
ax.plot(x,x+4,color='blue',lw=0.1)
plt.show()
```

```python
fig,ax=plt.subplots(dpi=150)
#line style
ax.plot(x,x+1,color='blue',lw=2,linestyle="-")
ax.plot(x,x+3,color='blue',lw=2,linestyle=':')
ax.plot(x,x+4,color='blue',lw=2,linestyle='-.')

line,=ax.plot(x,x+2,color='black',lw=1.5)
line.set_dashes([5,10,15,10])

line,=ax.plot(x,x+5,color='black',lw=1.5)
line.set_dashes([5,10,2,10])

line,=ax.plot(x,x+6,color='black',lw=1.5)
line.set_dashes([5,10,15,5])

line,=ax.plot(x,x+7,color='black',lw=1.5)
line.set_dashes([5,10,4,3])

plt.show()
```

```python
fig,ax=plt.subplots(dpi=150)
#line marker
ax.plot(x,x+1,color='blue',marker='o')
ax.plot(x,x+2,color='blue',marker='+')
ax.plot(x,x+3,color='blue',marker='s')
ax.plot(x,x+4,color='blue',marker='1')
ax.plot(x,x+5,color='blue',marker='o',markersize=2)
ax.plot(x,x+6,color='blue',marker='s',markerfacecolor='red')
ax.plot(x,x+7,color='blue',marker='o',markersize=5,markerfacecolor='green')
plt.show()
```

# other 2d graphs 

```python
n=np.array([0,1,2,3,4,5])
fig,axes=plt.subplots(1,4,figsize=(26,4))

axes[0].set_title('SCATTER PLOT')
axes[0].scatter(x,x+0.25*np.random.randn(len(x)),color='red')

axes[1].set_title('STEP PLOT')
axes[1].step(n,n**2,lw=2,color='black')

axes[2].set_title('BAR PLOT')
axes[2].bar(n,n**2,align='center',alpha=.5,color='green',width=.9)

axes[3].set_title("FILL BETWEEN")
axes[3].fill_between(x,x**2,x**3,color='blue',alpha=.5)
```

```python
#draw a radar chart
fig=plt.figure(figsize=(6,6))
ax=fig.add_axes([0.0,0.0,.6,.6],polar=True)
t=np.linspace(0,2*np.pi,100)
ax.plot(t,t*.5,color='blue',lw=2)
```

```python
#draw a histogram
n=np.random.randn(100000)
fig,axes=plt.subplots(1,2,figsize=(12,4))
axes[0].set_title('DEFAULT HISTOGRAM')
axes[0].hist(n,bins=20)

axes[1].set_title('CUMMULATIVE DERTAILED  HISTOGRAM')
axes[1].hist(n,cumulative=True,bins=50)
```

```python
#draw a contour map
import matplotlib.cm as cm
delta=0.025
x=np.arange(-3,3,delta)
y=np.arange(-2,2,delta)
X,Y=np.meshgrid(x,y)
z1=np.exp(X**2-Y**2)
z2=np.exp(-(X-1)**2-(Y-1)**2)
Z=(z1-z2)*2
print(X,Y,Z)
```

```python
fig,ax=plt.subplots()
cs=plt.contour(X,Y,Z)
ax.clabel(cs,inline=1,fontsize=10)
ax.set_title("CONTOUR MAP")
```
